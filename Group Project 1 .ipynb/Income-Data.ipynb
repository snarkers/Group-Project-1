```python
import pandas as pd
import numpy as np
import os
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
path = os.path.join("Texas Demographic Data", "Congressional District Data", "Texas County Income Data 2000-2018.csv")
df = pd.read_csv(path, skiprows = 4)
df.head(10)
```

```python
list(df)
```

```python
state_df = df.loc[df["GeoName"] == "Texas"]
state_df = state_df.set_index("Description")
state_df = state_df.drop(["GeoName"], axis = 1)
state_df
```

```python
texas_per_capita = state_df.loc["Per capita personal income (dollars) 2/"]
data = {"Texas Per-Capita Income $": texas_per_capita}
texas_per_capita_df = pd.DataFrame(data, columns = ["Texas Per-Capita Income $"])
texas_per_capita_df
```

```python
income_df = df.loc[df["Description"] == "Per capita personal income (dollars) 2/"]
income_df.set_index("GeoName")
```

```python
income_df = income_df.drop(["Description", "GeoFips", "LineCode"])
```

```python
income_df = income_df.set_index("GeoName")
```

```python
income_df = income_df.drop(["GeoFips", "LineCode", "Description"], axis = 1)
```

```python
income_df = income_df.transpose()
```

```python
income_df
```

```python
income_change = income_df["2018"] - income_df["2000"]
income_change
```

```python
change_df = pd.DataFrame({"Change in per-Capita Income": income_change}).reset_index()
change_df = change_df.set_index("GeoName")
change_df
```

```python
highest_growth = change_df["Change in per-Capita Income"].nlargest(10)
lowest_growth = change_df["Change in per-Capita Income"].nsmallest(10)
state_growth = change_df["Change in per-Capita Income"].iloc(0)


income_change_max_df = pd.DataFrame({"Counties with Highest per-Capita Income Growth": highest_growth} 
                                 ).reset_index()
income_change_min_df = pd.DataFrame({"Counties with Lowest per-Capita Income Growth": lowest_growth}).reset_index()
income_change_max_df
#state_growth_df = pd.DataFrame({"State per-Capita Income Growth": state_growth}).reset_index()
```

```python
#income_change_max_df = income_change_max_df.set_index("GeoName")
#income_change_min_df = income_change_min_df.set_index("GeoName")
```

```python
income_change_max_df.plot(kind = "bar", color = 'g')
income_change_min_df.plot(kind = "bar", color = 'r')
plt.show()
```

```python
#income_change_max_df.merge(income_change_min_df)
income_change_merge_df = income_change_max_df.merge(income_change_min_df, how='outer', left_index=False, right_index=False)
income_change_merge_df = income_change_merge_df.set_index("GeoName")
income_change_merge_df
```
